```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("car_crashes")
```

```python
sns.scatterplot(x = "alcohol", y = "speeding", data = df)
# Inference: We can see a slight linear positive relationship between alcohol and speeding.
# As the alcohol is increasing, speeding is increasing too.
```

```python
sns.scatterplot(x = "alcohol", y = "no_previous", data = df)
# Inference: We can see a linear positive relationship between alcohol and no previous accidents.
```

```python
sns.scatterplot(x = "speeding", y = "not_distracted", data = df)
# Inference: The points are all over the place.
# We can say there is no specific relationship between speeding and not distacted.
```

```python
sns.scatterplot(x = "alcohol", y = "total", data = df)
# Inference: We can see a linear positive relationship between alcohol and total accidents.
# People who drink more alcohol are tend to be in more accidents
```

```python
sns.lineplot(x = "alcohol", y = "total", data = df)
# Inference: We can see a linear positive relationship between alcohol and total accidents.
```

```python
sns.distplot(df["total"])
# Inference: We see 10 - 20 is the mean/average range where total accidents are more.
```

```python
sns.distplot(df["alcohol"])
# Inference: We see 10 - 20 is the mean/average range where total accidents are more.
```

```python
sns.relplot(x = "total", y = "alcohol", data = df, hue = "abbrev")
# Inference: The only categorical variable we have is abbrev, which is not related with the crashes.
# Hence there are no relationship in relplot.
```

For the same reason, barplot and countplot are more efficient on categorical variables.

As we have only one categorical variable - abbrev which has 51 different values, there can be no inference from the barplot and countplot drawn from it.

```python
sns.jointplot(x = "no_previous", y = "ins_premium", data = df)
# Inference: As we can see, there is no relation between the previous accidents and insurance premium money.
```

```python
sns.boxplot(df.total)
# Inference: No outliers present in the data. All the data for total car crashes are in a given range.
```

```python
sns.boxplot(df.alcohol)
# Inference: Few outliers present. This might mean, few of the people who has car crashes drank high amounts of alcohol.
```

```python
sns.boxplot(df.speeding)
# Inference: No outliers present in the data. All the speeds of car crashes are in a given range.
```
